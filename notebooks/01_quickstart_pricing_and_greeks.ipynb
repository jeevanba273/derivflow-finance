# Quickstart: Pricing, Greeks, and Implied Volatility

This notebook prices a European option, computes its Greeks, and recovers the implied volatility from the price.

```python
from derivflow import price_european_option, GreeksCalculator, implied_volatility

S, K, T, r, sigma = 100, 105, 0.25, 0.05, 0.20
```

## Price

```python
call = price_european_option(S, K, T, r, sigma, 'call')
put = price_european_option(S, K, T, r, sigma, 'put')
print(f'Call: {call:.4f}  Put: {put:.4f}')
```

## Greeks

```python
g = GreeksCalculator().calculate_greeks(S, K, T, r, sigma, 'call')
print(f'Delta {g.delta:.4f} Gamma {g.gamma:.4f} Vega {g.vega:.4f} Theta {g.theta:.4f} Rho {g.rho:.4f}')
```

## Implied volatility (inverting the price)

```python
iv = implied_volatility(call, S, K, T, r, 'call')
print(f'Implied vol: {iv:.4%} (input {sigma:.0%})')
```

## Dividend yield (generalized Black-Scholes-Merton)

```python
print('No dividend:', round(price_european_option(S, K, T, r, sigma, 'call'), 4))
print('3% dividend:', round(price_european_option(S, K, T, r, sigma, 'call', q=0.03), 4))
```
