# Volatility Surface and Implied Volatility

Build a sample volatility surface, interpolate it, and recover an implied volatility smile from option prices.

```python
from derivflow import create_sample_surface, price_european_option, implied_vol_from_chain
```

## Build and inspect the surface

```python
surface = create_sample_surface()
surface.build_surface()
stats = surface.surface_statistics()
print('points:', stats['num_points'])
print('vol range: %.1f%% - %.1f%%' % (stats['min_volatility']*100, stats['max_volatility']*100))
```

## Interpolate a point and extract a smile

```python
print('vol at K=100, T=0.25:', round(surface.interpolate(100, 0.25), 4))
smile = surface.get_smile(0.25, num_points=5)
for k, v in zip(smile['strikes'], smile['volatilities']):
    print(f'  K={k:.0f}: {v:.2%}')
```

## Recover an implied-volatility smile from prices

```python
chain = [
    {'strike': k, 'expiry': 0.5, 'option_type': 'call',
     'market_price': price_european_option(100, k, 0.5, 0.05, 0.20 + 0.10*abs(k-100)/100, 'call')}
    for k in (90, 95, 100, 105, 110)
]
for row in implied_vol_from_chain(chain, S=100, r=0.05):
    print(f"K={row['strike']:.0f}  IV={row['volatility']:.4f}")
```
