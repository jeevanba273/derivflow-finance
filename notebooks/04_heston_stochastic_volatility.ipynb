# Heston Stochastic Volatility Model

The Heston model adds a mean-reverting stochastic variance process. Here we set parameters (checking the Feller condition) and price by Monte Carlo.

```python
from derivflow import HestonModel, price_european_option
model = HestonModel()
model.set_parameters(v0=0.04, kappa=2.0, theta=0.04, sigma=0.3, rho=-0.7)
```

## Price by Monte Carlo and compare to Black-Scholes

```python
res = model.monte_carlo_price(100, 100, 1.0, 0.05, 'call', num_sims=50000, num_steps=200)
bs = price_european_option(100, 100, 1.0, 0.05, 0.2, 'call')
print('Heston MC call:', round(res.price, 4))
print('Black-Scholes :', round(bs, 4))
```

Note: with negative correlation (rho < 0) the Heston model produces a volatility skew, so its price differs from the flat-vol Black-Scholes value.
