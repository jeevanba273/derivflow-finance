# Portfolio Risk and Value-at-Risk

Build a multi-asset portfolio, aggregate its Greeks, and compute VaR.

```python
from derivflow import create_sample_portfolio
analyzer = create_sample_portfolio()
print('positions:', len(analyzer.positions))
print('value: %.2f' % analyzer.calculate_portfolio_value())
```

## Aggregate Greeks

```python
for name, value in analyzer.calculate_portfolio_greeks().items():
    print(f'  {name:6s}: {value:,.4f}')
```

## Parametric Value-at-Risk

```python
for cl in (0.95, 0.99):
    var, es = analyzer.calculate_var_parametric(cl)
    print(f'{cl:.0%} VaR: {var:,.2f}  Expected Shortfall: {es:,.2f}')
```
