# Market Data and Dashboards

Live market data uses yfinance and needs an internet connection. This notebook degrades gracefully to sample data when offline, then builds interactive Plotly dashboards (which need no network).

```python
from derivflow import AdvancedMarketData, DerivativesDashboard, create_sample_surface
md = AdvancedMarketData()
is_open, status = md.is_market_open()
print('market status:', status)
```

## Live price with an offline fallback

```python
try:
    spot, _ = md.get_current_price('AAPL')
    print('AAPL spot:', round(spot, 2))
except Exception:
    spot = 150.0
    print('offline - using sample spot', spot)
```

## Dashboards (offline)

Each `plot_*` method returns a Plotly figure; call `fig.show()` to render it.

```python
dash = DerivativesDashboard()
payoff = dash.plot_option_payoff(100, 5.0, 'call')
greeks = dash.plot_greeks_dashboard(100, 100, 0.25, 0.05, 0.2, 'call')
print('payoff figure:', type(payoff).__name__)
print('greeks figure:', type(greeks).__name__)
# payoff.show()
# greeks.show()
```
